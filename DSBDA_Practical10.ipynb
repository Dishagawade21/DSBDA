```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris_df = sns.load_dataset('iris')
print("First five rows of the dataset:")
print(iris_df.head())
print("\nDataset information:")
iris_df.info()
print("\nSummary statistics:")
print(iris_df.describe())
```

```python
numeric_features = ['sepal_length', 'sepal_width', 'petal_length',
'petal_width']
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()
for i, feature in enumerate(numeric_features):
  sns.histplot(data=iris_df, x=feature, kde=True, ax=axs[i],
color='skyblue', edgecolor='black')
axs[i].set_title(f'Histogram and KDE of {feature}', fontsize=12)
axs[i].set_xlabel(feature, fontsize=10)
axs[i].set_ylabel("Frequency", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()
for i, feature in enumerate(numeric_features):
  sns.boxplot(y=iris_df[feature], ax=axs[i], color='lightgreen')
axs[i].set_title(f'Box Plot of {feature}', fontsize=12)
axs[i].set_ylabel(feature, fontsize=10)
plt.tight_layout()
plt.show()
```
